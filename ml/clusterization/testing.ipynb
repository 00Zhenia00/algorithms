---
Agglomerative algorithm
---

### Imports

```python
from sklearn.cluster import AgglomerativeClustering
from Agglomerative import Agglomerative
```

### First check

```python
# from sklearn tutorial
data = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [4, 2],
    [4, 4],
    [4, 0]
])

my_model = Agglomerative(n_clusters=2, metric='euclidean', linkage='single')
sk_model = AgglomerativeClustering(n_clusters=2, linkage='single')
my_model.fit(data)
sk_model.fit(data)
print(my_model.labels)
print(sk_model.labels_)
```

### Second check

```python
datasets_list = generate_data()

models = [
    (
        "CustomAgglomerative",
        lambda params: Agglomerative(
            n_clusters=params["n_clusters"], metric="euclidean", linkage="ward"
        ),
    ),
    (
        "sklearn",
        lambda params: AgglomerativeClustering(
            n_clusters=params["n_clusters"], linkage="ward"
        ),
    ),
]

params = {
    "noisy_circles": {"n_clusters": 2},
    "noisy_moons": {"n_clusters": 2},
    "varied": {"n_clusters": 3},
    "aniso": {"n_clusters": 3},
    "blobs": {"n_clusters": 3},
    "no_structure": {"n_clusters": 3},
}

plot_models_per_dataset(models, datasets_list, params)

# compare_model_plots(models, datasets_list, params)
```

---
K-Means
---

### Imports

```python
from sklearn.cluster import KMeans as skKMeans
from KMeans import KMeans
```

### First check

```python
# from sklearn tutorial
X = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]])

my_kmeans = KMeans(
    n_clusters=2, max_iter=10, tol=1e-4, init="k-means++", random_state=170
).fit(X)
sk_kmeans = skKMeans(n_clusters=2, random_state=170, init="k-means++").fit(X)

print(sk_kmeans.labels_)
print(my_kmeans.labels)
```

### Second check

```python
datasets_list = generate_data()

models = [
    ("CustomKMeans", lambda params: KMeans(n_clusters=params["n_clusters"], init="k-means++", random_state=170)),
    ("sklearn", lambda params: skKMeans(n_clusters=params["n_clusters"], init="k-means++", random_state=170)),
]

params = {
    "noisy_circles": {"n_clusters": 2},
    "noisy_moons": {"n_clusters": 2},
    "varied": {"n_clusters": 3},
    "aniso": {"n_clusters": 3},
    "blobs": {"n_clusters": 3},
    "no_structure": {"n_clusters": 3},
}

# compare_model_plots(models, datasets_list, params)

plot_models_per_dataset(models, datasets_list, params)
```

DBSCAN

### Imports

```python
from DBSCAN import DBSCAN
from sklearn.cluster import DBSCAN as skDBSCAN
```

```python
# from sklearn tutorial
X = np.array([[1, 2], [2, 2], [2, 3], [8, 7], [8, 8], [25, 80]])

my_dbscan = DBSCAN(eps=3, min_samples=2).fit(X)
sk_dbscan = skDBSCAN(eps=3, min_samples=2).fit(X)

print(sk_dbscan.labels_)
print(my_dbscan.labels)
```

### Second check

```python
datasets_list = generate_data()

models = [
    ("CustomDBSCAN", lambda params: DBSCAN(eps=params["eps"], min_samples=params["min_samples"])),
    ("sklearn", lambda params: skDBSCAN(eps=params["eps"], min_samples=params["min_samples"])),
]

params = {
    "noisy_circles": {"eps": 0.3, "min_samples": 7},
    "noisy_moons": {"eps": 0.3, "min_samples": 7},
    "varied": {"eps": 0.3, "min_samples": 7},
    "aniso": {"eps": 0.15, "min_samples": 7},
    "blobs": {"eps": 0.3, "min_samples": 7},
    "no_structure": {"eps": 0.3, "min_samples": 7},
}

# compare_model_plots(models, datasets_list, params)

plot_models_per_dataset(models, datasets_list, params)
```
